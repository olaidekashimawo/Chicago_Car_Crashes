```python
def get_time_of_day(crash_hour):
    if crash_hour >= 0 and crash_hour <= 6:
        return 'Early Morning'
    elif crash_hour > 6 and crash_hour <= 12:
        return 'Morning'
    elif crash_hour > 12 and crash_hour <= 18:
        return 'Afternoon'
    else:
        return 'Evening'
df['time_of_day'] = pd.x=df['crash_hour'].apply(lambda x: get_time_of_day(x))
```

## Functions for Modeling
* Input a model, training data and test data to return sklearn metrics
        - Classification Report for training and test
        - Confusion Matrix for training and test
        - ROC Curve for training and test
        - adding legend


```python
def evaluate_classification(model, X_train_tf, X_test_tf, y_train, y_test, classes=None, normalize='true', cmap='Blues_r', label=''):
    """Input a model, training data and test data to return sklearn metrics
        - Classification Report for training and test
        - Confusion Matrix for training and test
        - ROC Curve for training and test

        Args:
        model: instantiated classifier (ex: LogisticRegression = log_reg)
        X_train_tf: train_test_split & preprocessed
        X_test_tf: train_test_split & preprocessed
        y_train: train_test_split
        y_test: train_test_split
        label: optional label for type of test

        Modified Function, Citation:
        https://github.com/jirvingphd/fsds_070620_FT_cohort_notes/blob/master/Mod_3/
        mod_3_classification_project/bonus%20notebooks/classification_project_OH1_pipelines_SHAP%20.ipynb

    """
    # retrieve predictions for train and test data
    y_pred_train = model.predict(X_train_tf)
    y_pred_test = model.predict(X_test_tf)

    # print training classification report
    header = label + " CLASSIFICATION REPORT TRAINING "
    dashes = "---" * 20
    print(dashes, header, dashes, sep='\n')
    print(classification_report(y_train, y_pred_train, target_names=classes))

    # display training figures
    fig, axes = plt.subplots(figsize=(10, 4), ncols=2)

    # confusion matrix
    plot_confusion_matrix(model, X_train_tf, y_train, display_labels=['Not Injured','Injured'], normalize=normalize,
                          cmap=cmap, ax=axes[0])
    axes[0].set(title='Confusion Matrix Training')

    # plot ROC curve
    plot_roc_curve(model, X_train_tf, y_train, ax=axes[1])
    roc = axes[1]
    roc.legend(title='Training', loc='lower right')
    roc.plot([0, 1], [0, 1], ls=':')
    roc.grid()
    roc.set_title('Receiving Operator Characteristic Training')

    # add legend
    handles, labels = roc.get_legend_handles_labels()
    roc.legend(handles=handles, labels=labels, title='Training')

    plt.show()

    # print testing classification report
    header_ = label + " CLASSIFICATION REPORT TESTING "
    print(dashes, header_, dashes, sep='\n')
    print(classification_report(y_test, y_pred_test, target_names=classes))

    # display testing figures
    fig, axes = plt.subplots(figsize=(10, 4), ncols=2)

    # confusion matrix
    plot_confusion_matrix(model, X_test_tf, y_test, display_labels=['Non Injured','Injured'], normalize=normalize,
                          cmap=cmap, ax=axes[0])
    axes[0].set(title='Confusion Matrix Testing')

    # plot ROC curve
    plot_roc_curve(model, X_test_tf, y_test, ax=axes[1])
    roc = axes[1]
    roc.legend(title='Testing', loc='lower right')
    roc.plot([0, 1], [0, 1], ls=':')
    roc.grid()
    roc.set_title('Receiving Operator Characteristic Test')

    # add legend
    handles, labels = roc.get_legend_handles_labels()
    roc.legend(handles=handles, labels=labels, title='Testing')

    plt.show()

    
```

## Class Imbalance Using SMOTE (Synthetic Minority Over-sampling Technique)
Since we have an imbalance dataset SMOTE was used to generate new synthetic samples for the minority class based on the existing samples. It does this by selecting a random sample from the minority class and finding its k nearest neighbors. It then generates a new sample by randomly selecting one of the k neighbors and creating a synthetic sample along the line connecting the original sample and the selected neighbor. This process is repeated until the desired balance between the two classes is achieved.

A confusion matrix typically has four components:

* True Positive (TP): The number of samples that were correctly predicted as positive.
* False Positive (FP): The number of samples that were incorrectly predicted as positive.
* False Negative (FN): The number of samples that were incorrectly predicted as negative.
* True Negative (TN): The number of samples that were correctly predicted as negative.

Precision, recall, and F1-score are commonly used metrics to evaluate the performance of a classification model based on its confusion matrix. Here's a brief description of each metric:

* Precision: Precision is the fraction of true positives among all positive predictions. It measures the model's ability to correctly identify positive samples. A high precision means that the model rarely makes false positive predictions.
* Recall: Recall is the fraction of true positives among all actual positive samples. It measures the model's ability to correctly identify all positive samples, regardless of whether some are misclassified as negative. A high recall means that the model rarely makes false negative predictions.
* F1-score: The F1-score is the harmonic mean of precision and recall, and it balances both metrics. It ranges from 0 to 1, where a higher score indicates better performance.

We are calculating these metrics using the classification_report()

SHAP (SHapley Additive exPlanations) is a unified framework for explaining the output of any machine learning model. It provides a way to understand the contribution of each feature or variable to the final prediction by measuring their impact on the model's output.

SHAP values are based on the concept of Shapley values, which is a method from cooperative game theory that assigns a value to each player based on their contribution to a coalition. In the case of machine learning models, each feature or variable is treated as a player, and the prediction score is treated as the value of the coalition.



```python
X_shap = shap.sample(X_train_tf)
explainer = shap.TreeExplainer(best_clf)
shap_values = explainer.shap_values(X_shap, check_additivity=False)
shap.summary_plot(shap_values, X_shap, plot_type="bar")
```
